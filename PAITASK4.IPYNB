#### Task 4 Implement the N-Queens Problem (Dynami

```python
def solve_n_queens(n):
    solutions = []
    
    cols = set()
    diag1 = set()   
    diag2 = set()   
    
    board = [["." for _ in range(n)] for _ in range(n)]
    
    def backtrack(row):
        if row == n:
           
            solutions.append(["".join(r) for r in board])
            return
        
        for col in range(n):
            if col in cols or (row - col) in diag1 or (row + col) in diag2:
                continue
            
        
            board[row][col] = "Q"
            cols.add(col)
            diag1.add(row - col)
            diag2.add(row + col)
            
            print(f"Placed Q at ({row}, {col})")
            
            backtrack(row + 1)
           
            board[row][col] = "."
            cols.remove(col)
            diag1.remove(row - col)
            diag2.remove(row + col)
            
            print(f"Removed Q from ({row}, {col})")
    
    backtrack(0)
    return solutions



n = 4
solutions = solve_n_queens(n)

print("\nTotal Solutions:", len(solutions))
for sol in solutions:
    for row in sol:
        print(row)
    print()
```
