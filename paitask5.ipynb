#### Read & Display Your Image

```python
image_path = r"C:\Users\HP\Downloads\closeup-shot-beautiful-butterfly-with-interesting-textures-orange-petaled-flower.jpg"


img = cv2.imread(image_path)

if img is None:
    print("Image not found! Check the path.")
else:
    print("Image loaded successfully!")
    # Convert BGR to RGB for correct colors
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # Display
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show()
```

##### Save the Image

```python

cv2.imwrite("butterfly_copy.jpg", img)

print(" Image saved as 'butterfly_copy.jpg' in the notebook folder")
```

#### Resize the Image

```python

resized = cv2.resize(img, (300, 300))


from matplotlib import pyplot as plt
resized_rgb = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)
plt.imshow(resized_rgb)
plt.axis('off')
plt.show()

print("Image resized to 300x300 pixels")
```

#### Crop the Image

```python

cropped = img[100:400, 150:450]  
from matplotlib import pyplot as plt
cropped_rgb = cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)
plt.imshow(cropped_rgb)
plt.axis('off')
plt.show()

print("Image cropped successfully")
```

#### Convert to Grayscale

```python

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


from matplotlib import pyplot as plt
plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.show()

print(" Image converted to grayscale")
```

#### Draw Shapes & Text

```python

img_draw = img.copy()


cv2.rectangle(img_draw, (50, 50), (300, 300), (0, 255, 0), 3)


cv2.circle(img_draw, (350, 200), 80, (255, 0, 0), 5)


cv2.line(img_draw, (0, 0), (400, 400), (0, 0, 255), 3)


cv2.putText(img_draw, "Butterfly", (100, 380), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,0), 2)


from matplotlib import pyplot as plt
img_draw_rgb = cv2.cvtColor(img_draw, cv2.COLOR_BGR2RGB)
plt.imshow(img_draw_rgb)
plt.axis('off')
plt.show()

print(" Shapes and text drawn on the image")
```

#### Image Thresholding

```python

_, th = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
from matplotlib import pyplot as plt
plt.imshow(th, cmap='gray')
plt.axis('off')
plt.show()

print(" Thresholding applied")
```

#### Edge Detection

```python

edges = cv2.Canny(gray, 100, 200)  
from matplotlib import pyplot as plt
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()

print(" Edge detection applied")
```

#### Contours & Shape Detection

```python

contours, _ = cv2.findContours(th, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

img_contours = img.copy()

cv2.drawContours(img_contours, contours, -1, (0, 0, 255), 2)

from matplotlib import pyplot as plt
img_contours_rgb = cv2.cvtColor(img_contours, cv2.COLOR_BGR2RGB)
plt.imshow(img_contours_rgb)
plt.axis('off')
plt.show()

print(f" {len(contours)} contours found and drawn on the image")
```

#### Face Detection (Template)

```python

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")


gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.3, minNeighbors=5)


img_faces = img.copy()
for (x, y, w, h) in faces:
    cv2.rectangle(img_faces, (x, y), (x + w, y + h), (255, 0, 0), 2)


from matplotlib import pyplot as plt
img_faces_rgb = cv2.cvtColor(img_faces, cv2.COLOR_BGR2RGB)
plt.imshow(img_faces_rgb)
plt.axis('off')
plt.show()

print(f" Number of faces detected: {len(faces)}")
```
